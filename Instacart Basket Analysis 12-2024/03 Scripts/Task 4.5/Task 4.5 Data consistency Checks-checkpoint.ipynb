# Addressing missing values and outliers

```python
# Creating a df_prods without the missing values
df_prods_clean = df_prods[df_prods['product_name'].isnull() == False]
```

```python
df_prods_clean.shape
```

```python
# Identifying products with potential price outliers.
df_prods[df_prods['prices'] > df_prods['prices'].mean() + (2 * df_prods['prices'].std())]
```

```python
# Replacing outlier values 99999.00 and 14900.00 with the mean value of 9.994
df_prods_clean = df_prods.replace({'prices':{99999.0:9.994, 14900.0:9.994}})
```

```python
# Checking df_prods_clean statistics with replaced outliers
df_prods_clean.describe()
```

# Explanation on the products df statistics
The statiscal analysis on the products dataframe indicates 2 extreme values in the prices column. It is highly unlikely that Lowfat 2% milkfat Cheese and 2% reduced fat milk costs 99999 and 14900 respectively as it's been proven when looking for outliers in the price column. These 2 extreme values are far more than the upper bound limit of (mean + 2 * Standard deviations) and i can comfortably consider them as outliers. Replacing the outlier values with the mean value (I chosed mean instead of median value) reduces the price's mean value from 9.994 to 7.68
