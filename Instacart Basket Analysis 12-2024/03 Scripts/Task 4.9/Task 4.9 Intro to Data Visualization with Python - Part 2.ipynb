## Step 7
### Exploring connection between age and spending power(income) using a scatterplot

```python
# Using the 'small' sample to create a subset with needed columns
df_4 = small[['age', 'income']]
```

```python
# Creating a scatterplot
scatterplot = sns.scatterplot(data = df_4, x = 'age', y = 'income')
```

The plot suggests that between the ages of 16 and 40 the average income is slightly above 200.000 with pockets of that age group earning as high as 400.000. On the other hand, customers 42 years and above have an average income of above 300.000 with some earning as much as close to 600.000. The denser the plots, the more the people in that age group relative to the income bracket and as it can be seen on the plot, customers from 42 years have a considerably higher income and are likely to spend more on purchases.
