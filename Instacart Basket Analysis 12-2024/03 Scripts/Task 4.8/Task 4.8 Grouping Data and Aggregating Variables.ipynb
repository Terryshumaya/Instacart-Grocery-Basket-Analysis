### Creating a new column max_order

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['number_of_order'].transform(np.max)
```

```python
ords_prods_merge.head(20)
```

# Step 4
### Assigning loyalty flag

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking loyalty flag counts
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Checking loyalty flag with selected columns
ords_prods_merge[['user_id', 'number_of_order', 'max_order', 'loyalty_flag']].head(30)
```

# Step 5

```python
# Determining the 'min', 'max' and 'mean' values of each loyal flag
ords_prods_merge.groupby(['loyalty_flag']).agg({'prices':['mean', 'min', 'max']})
```

Simply looking at the values one could doubt the results because new customer has a higher average price than the loyal customer with both having the same min and max prices. It should however be noted that, despite loyal customer having a slightly lower average than both the regular and new customers, it has more orders(40 and above) making its price average to drop within the min and max range of 1.0 to 25.0. 

# Step 6

```python
# Creating column for average spender
ords_prods_merge['average_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Creating flag for low and high spender
ords_prods_merge.loc[ords_prods_merge['average_spend'] < 10, 'spender_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spend'] > 10, 'spender_flag'] = 'High spender'
```

```python
# Checking spender flag counts
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
# Checking average_spend and spender_flag columns 
pd.options.display.max_rows = None
```

```python
ords_prods_merge[['user_id', 'prices', 'average_spend', 'spender_flag']].head(60)
```

# Step 7

```python
# Creating a frequent-users column
ords_prods_merge['Customer_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Creating a flag on order frequency
ords_prods_merge.loc[ords_prods_merge['Customer_frequency'] > 20, 'frequency_flag'] = 'Non frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['Customer_frequency'] > 10) & (ords_prods_merge['Customer_frequency'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['Customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer' 
```

```python
# Checking frequency flags counts
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Checking newly created customer frequency and frequency_flag colunms
ords_prods_merge[['user_id', 'days_since_prior_order', 'Customer_frequency', 'frequency_flag']].head(100)
```
